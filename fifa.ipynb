```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa=pd.read_csv('fifa_data.csv')
```

```python
fifa.head()
```

```python
fifa.shape
```

```python
#Print all column names
for col in fifa.columns:
    print(col)
```

```python
#Show all countries and how many players for each country
fifa['nationality'].value_counts()
```

```python
#Top 10 countries with most players
fifa['nationality'].value_counts()[0:10]
```

```python
#Top 5 countries with most players
fifa['nationality'].value_counts()[0:5]
```

```python
#SBar chart showing top 5 countries with most players
plt.figure(figsize=(7,7))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()), list(fifa['nationality'].value_counts()[0:5]), color="teal")
plt.title("Top 5 countries where most players belong")
plt.xlabel("Name of country")
plt.ylabel("Number of players")
plt.show()
```

```python
#Create databash showing just player name and what they earn
player_salary = fifa[['short_name', 'wage_eur']]
```

```python
player_salary.head()
```

```python
#Sort data in descending order
player_salary = player_salary.sort_values(by=('wage_eur'),ascending=False)
```

```python
player_salary.head()
```

```python
#Create a bar chart showing top 5 players and what they earn
plt.figure(figsize=(7,7))
plt.bar(list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5], color=["blue", "red", "yellow", "green", "orange"])
plt.title("Top 5 highest earners")
plt.xlabel("Player name")
plt.ylabel("Salary earned")
plt.xticks(rotation=45)
plt.show()
```

```python
fifa['nationality']=="Germany"
```

```python
#Create database only including info for Germany
Germany=fifa[fifa['nationality']=="Germany"]
Germany.head(10)
```

```python
#Show top 10 players from tallest to shortest
Germany.sort_values(by=['height_cm'], ascending=False).head(10)
```

```python
#Show top 10 players from heaviest to lightest
Germany.sort_values(by=['weight_kg'], ascending=False).head(10)
```

```python
#Show top 10 players from paid most to least
germany_wages = Germany.sort_values(by=['wage_eur'], ascending=False)
germany_wages.head(10)
```

```python
plt.figure(figsize=(7,7))
plt.barh(germany_wages['short_name'][0:10], germany_wages['wage_eur'][0:10], color = "teal")
plt.title("Top 10 Salaries for Germany")
plt.ylabel("Player Name")
plt.xlabel("Salary")
plt.xticks(rotation = 45)
plt.show()
```

```python
#German player name and earning in descending order
Germany[['short_name', 'wage_eur']].sort_values(by=['wage_eur'],ascending=False).head()
```

```python
#Create datafram of player name and shooting score
player_shooting=fifa[['short_name', 'shooting']]
```

```python
#Top shooters in descending order
player_shooting.sort_values(by=['shooting'],ascending=False).head()
```

```python
#Create database showing only players who play for Real Madrid
real_madrid = fifa[fifa['club']=='Real Madrid']
```

```python
#Sort by wages in desending order
real_madrid_wages = real_madrid.sort_values(by=['wage_eur'], ascending=False)
real_madrid_wages.head()
```

```python
#Bar chart showing highest paid Real Madrid players
plt.figure(figsize=(7,7))
plt.barh(real_madrid_wages['short_name'][0:10], real_madrid_wages['wage_eur'][0:10], color = "#4CAF50")
plt.title("Top 10 Salaries for Real Madrid")
plt.ylabel("Player Name")
plt.xlabel("Salary")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Sort shooting in desending order
real_madrid.sort_values(by=['shooting'], ascending=False).head()
```

```python
#Sort defending in desending order
real_madrid.sort_values(by=['defending'], ascending=False).head()
```

```python
#Nationality count in Real Madrid
real_madrid['nationality'].value_counts()
```

```python
#create pie chart showing nationality
plt.figure(figsize=(7,7))
plt.pie(list(real_madrid['nationality'].value_counts()), labels=list(real_madrid['nationality'].value_counts().keys()))
plt.show()
```

```python
#Plot scatter chart showing relationship of shooting and salary earned
plt.figure(figsize=(7,7))
plt.scatter(fifa['shooting'], fifa['wage_eur'], color="hotpink")
plt.title('Shooting vs Salary')
plt.xlabel('Shooting score')
plt.ylabel('Salary')
plt.show()
```

```python
#Plot scatter chart showing relationship of defending and salary earned
plt.figure(figsize=(7,7))
plt.scatter(fifa['defending'], fifa['wage_eur'])
plt.title('Defending vs Salary')
plt.xlabel('Defending score')
plt.ylabel('Salary')
plt.show()
```
